Langkah 1: Impor Pustaka yang Dibutuhkan

```python
import numpy as np
import matplotlib.pyplot as plt
```

Langkah 2: Membuat Data untuk Plot

```python
x = np.linspace(0, 10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)
```

Langkah 3: Menggunakan Berbagai Gaya Plot

```python
# Menampilkan gaya yang tersedia
print(plt.style.available)
plt.style.use('seaborn-whitegrid')
```

Langkah 4: Membuat Multiple Subplots dan Menyesuaikannya

```python
fig, axes = plt.subplots(2, 1, figsize=(10, 6))

# Subplot pertama
axes[0].plot(x, y1, color='blue', label='sin(x)')
axes[0].set_title('Plot of sin(x)')
axes[0].set_ylabel('sin(x)')
axes[0].legend()

# Subplot kedua
axes[1].plot(x, y2, color='red', label='cos(x)')
axes[1].set_title('Plot of cos(x)')
axes[1].set_xlabel('x')
axes[1].set_ylabel('cos(x)')
axes[1].legend()

# Penyesuaian layout
plt.tight_layout()
plt.show()
```
